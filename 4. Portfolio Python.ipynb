# ETL - adecuación de la estructura de la tabla

## Importamos la librería de trabajo

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Conectamos y mostramos el DataFrame

```python
file_path = r"C:\Users\danny\OneDrive - BigCsv\2. Formaciones\3. Pregrabadas\5. Portfolio, CV y Linkedin\2. Portfolios\1. PortFolio Excel\GranSantiagoMercadoWeb_inmobiliaria.xlsx"
df = pd.read_excel(file_path, sheet_name="Cifras Inmobiliarias Santiago")
df
```

## Empezamos con el proceso ETL

```python
# Creamos listas para cada columna de nuestro DataFrame
anio = []
meses = []
departamento_stock = []
departamento_ventas = []
departamento_meses = []
casas_stock = []
casas_ventas = []
casas_meses = []
viviendas_stock = []
viviendas_ventas = []
viviendas_meses = []
```

```python
#Creamos una lista de columnas para utilizarlas en nuestro bucle
df.columns
```

```python
# Rellenamos las listas con los datos del DataFrame importado
for index, row in df.iterrows():
    anio.append(row['MERCADO GRAN SANTIAGO'])
    meses.append(row['Unnamed: 1'])
    departamento_stock.append(row['Departamentos'])    
    departamento_ventas.append(row['Unnamed: 3'])    
    departamento_meses.append(row['Unnamed: 4'])    
    casas_stock.append(row['Casas'])    
    casas_ventas.append(row['Unnamed: 6'])    
    casas_meses.append(row['Unnamed: 7'])    
    viviendas_stock.append(row['Viviendas'])    
    viviendas_ventas.append(row['Unnamed: 9'])    
    viviendas_meses.append(row['Unnamed: 10'])
```

```python
# Creación del Dataframe nuevo para ajustar nuestra estructura tabular
df_t = pd.DataFrame({
    'Anio': anio,
    'Mes': meses,
    'Departamento_Stock': departamento_stock,
    'Departamento_Ventas': departamento_ventas,
    'Departamento_Meses': departamento_meses,
    'Casas_Stock': casas_stock,
    'Casas_Ventas': casas_ventas,
    'Casas_Meses': casas_meses,
    'Viviendas_Stock': viviendas_stock,
    'Viviendas_Ventas': viviendas_ventas,
    'Viviendas_Meses': viviendas_meses
})
# Mostramos el DataFrame
df_t.tail(10)
```

```python
# Eliminamos las filas desde donde aparece 'fuente' en la columna 'Año' hacia abajo ya que esta es nuestra referencia final de filas.
index_fuente = df_t[df_t['Anio'].str.contains('FUENTE', case=False, na=False)].index
if not index_fuente.empty:
    df_t = df_t[:index_fuente[0]]
```

```python
#Mostramos el dataframe
df_t.tail(10)
```

```python
# eliminamos filas que tengan NaN en toda la fila
df_t = df_t.dropna(how='all')

#rellenamos hacia abajo en el campo anio
df_t['Anio'] = df_t['Anio'].ffill()

#Mostramos el DataFrame
df_t
```

```python
# Utilizar melt para reorganizar el DataFrame
df_melted = pd.melt(df_t, id_vars=["Anio", "Mes"], 
                    value_vars=["Departamento_Stock", "Departamento_Ventas", "Departamento_Meses",
                                "Casas_Stock", "Casas_Ventas", "Casas_Meses",
                                "Viviendas_Stock", "Viviendas_Ventas", "Viviendas_Meses"],
                    var_name="Inmuebles_Variable", value_name="Value")
```

```python
df_melted 
```

```python
# Separar la columna 'Inmuebles_Variable' en 'Inmuebles' y 'Variable'
df_melted[['Inmuebles', 'Variable']] = df_melted['Inmuebles_Variable'].str.split('_', expand=True)
```

```python
df_melted
```

```python
# Pivotar de nuevo el DataFrame para tener las columnas 'Stock', 'Ventas' y 'Meses'
df_pivoted = df_melted.pivot_table(index=["Anio", "Mes", "Inmuebles"], columns="Variable", values="Value").reset_index()
```

```python
df_pivoted
```

```python
# Ordenar las columnas
df_final = df_pivoted[['Anio', 'Mes', 'Inmuebles', 'Stock', 'Ventas', 'Meses']]

# Mostrar el DataFrame reorganizado
df_final
```

```python
#verifico que la suma me coincida
ventas_totales = df_final['Ventas'].sum()
print("Total Ventas:", ventas_totales)
```

```python
# Crear un diccionario para mapear los nombres de los meses a números
mes_map = {
    "Ene": 1, "Feb": 2, "Mar": 3, "Abr": 4, "May": 5, "Jun": 6,
    "Jul": 7, "Ago": 8, "Sep": 9, "Oct": 10, "Nov": 11, "Dic": 12
}
```

```python
# Convertir la columna 'Mes' a números usando el diccionario
df_final['Mes'] = df_final['Mes'].map(mes_map)
```

```python
# Renombrar las columnas a 'year', 'month' y 'day'
df_final = df_final.rename(columns={'Anio': 'year', 'Mes': 'month'})
df_final['day'] = 1
```

```python
df_final
```

```python
# Crear una nueva columna de tipo fecha combinando 'year', 'month' y 'day'
df_final['Fecha'] = pd.to_datetime(df_final[['year', 'month', 'day']])
```

```python
df_final
```

```python
# Eliminar las columnas originales 'year', 'month' y 'day' si ya no son necesarias
df_final = df_final.drop(columns=['year', 'month', 'day'])

# Reorganizar las columnas para que 'Fecha' esté al inicio
df_final = df_final[['Fecha', 'Inmuebles', 'Stock', 'Ventas', 'Meses']]
```

# EDA Análisis exploratorio

```python
# Convertir las columnas 'Stock' y 'Ventas' a enteros para poder usarlos en el método describe
df_final['Stock'] = pd.to_numeric(df_final['Stock'], errors='coerce').astype('int64')
df_final['Ventas'] = pd.to_numeric(df_final['Ventas'], errors='coerce').astype('int64')

# Convertir la columna 'Meses' a valores decimales (float)
df_final['Meses'] = pd.to_numeric(df_final['Meses'], errors='coerce')

# Verificar la conversión
df_final.info()
```

```python
df_final.describe()
```

```python
df_final.isnull().sum()
```

```python
# Distribución de las Ventas por Tipo de Inmueble
plt.figure(figsize=(12, 6))
sns.boxplot(x='Inmuebles', y='Ventas', data=df_final)
plt.title('Distribución de las Ventas por Tipo de Inmueble')
plt.show()
```

```python
# Evolución de las Ventas a lo Largo del Tiempo
plt.figure(figsize=(12, 6))
sns.lineplot(x='Fecha', y='Ventas', hue='Inmuebles', data=df_final)
plt.title('Evolución de las Ventas a lo Largo del Tiempo')
plt.show()
```

```python
# Seleccionar solo las columnas numéricas para la matriz de correlación
numeric_df = df_final.select_dtypes(include=['int64', 'float64'])

# Matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Evolución de Stock y Ventas a lo largo del tiempo
plt.figure(figsize=(14, 7))

# Evolución de Stock
plt.subplot(2, 1, 1)
sns.lineplot(x='Fecha', y='Stock', hue='Inmuebles', data=df_final)
plt.title('Evolución del Stock a lo Largo del Tiempo')
plt.xlabel('')
plt.ylabel('Stock')

# Evolución de Ventas
plt.subplot(2, 1, 2)
sns.lineplot(x='Fecha', y='Ventas', hue='Inmuebles', data=df_final)
plt.title('Evolución de las Ventas a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ventas')

plt.tight_layout()
plt.show()
```
